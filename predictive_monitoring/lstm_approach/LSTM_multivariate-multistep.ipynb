# LSTM multi-step-ahead prediction

This notebook explores how to predict the previously defined high-level metric *Efficiency*.
The code wraps up the data processing, the model training and testing phases, and the model export. 

Except for the model's implementation details, this notebooks runs the code with some fixed parameters/targets:
- Lookback: 24 sampling steps (2 hours history)
- Batch size: 72 elements (inside this batch, the long memory is kept)
- Prediction of the three next *Efficiency* sampling steps, i.e., next 5 mins, next 10 mins, next 15 mins

## Data preprocessing

```python
JOBID = 3418339

columns_to_consider = [
    "end time",
    "CPU rate",
    "canonical memory usage",
    "assigned memory usage",
    "unmapped page cache",
    "total page cache",
    "maximum memory usage",
    "disk I/O time",
    "local disk space usage",
    "maximum CPU rate",
    "maximum disk IO time",
    "cycles per instruction",
    "memory accesses per instruction",
    "Efficiency",  # target metric
]

readings_df = load_data(
    "../data/task-usage_job-ID-%i_total.csv" % JOBID, columns_to_consider
)
```

```python
readings_df.head(10)
```

### Generate timestamp to interpolate NaN

In this part, we are taking the end time of the observation period, and we are converting them to datetime values.
As start time, we consider the 1st of May, 2011, PT. We add this value to the end time values.

```python
readings_df["datetime"] = timestamps_readings = [
    pd.Timestamp(int(t / 1000000) + 1304233200, tz="US/Pacific", unit="s")
    for t in readings_df["end time"].values
]
```

Here, we are setting the datetime column as index,as needed by the interpolate function from pandas

```python
readings_df.set_index("datetime", inplace=True)
```

```python
readings_df.head()
```

Use interpolation to remove nan. Sources:
- https://stackoverflow.com/questions/49308530/missing-values-in-time-series-in-python
- https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.interpolate.html



```python
readings_df.interpolate(method="time", inplace=True)
```

### Aggregate data

```python
def q95(x):
    return x.quantile(0.95)

def q75(x):
    return x.quantile(0.75)

def q25(x):
    return x.quantile(0.25)


```

```python
readings_df_x = readings_df.groupby('end time').agg([('mean', 'mean'), ('median', 'median'), ('min', 'min'), ('max','max'), ('q25', q25), ('q75', q75), ('q95', q95), ('std', 'std')])
```

```python
readings_df_x
```

```python
pyplot.figure(figsize=(21, 9))

readings_df_x["Efficiency"]['mean'].plot()
```

```python
readings_df = readings_df_x[columns_to_consider[1:-1]]
readings_df["Efficiency"] = readings_df_x["Efficiency"]["mean"].values
readings_df
```

### Scale values

```python
values = readings_df.values

scaled, scaler = scale_values(values)
```

### Create train-validation-test sets

```python
n_train = int(scaled.shape[0] * 0.6)
n_validation = int(scaled.shape[0] * 0.2)
train = scaled[:n_train, :]
validation = scaled[n_train : (n_train + n_validation), :]
test = scaled[(n_train + n_validation) :, :]

print("Train shape:", train.shape)
print("Validation shape:", validation.shape)
print("Test shape:", test.shape)
```

### Reshape the input data without Efficiency - __no_eff__

Reshape data, keeping out efficiency, that is our *target metric*, and according to the defined lookback and lookahead values, namely:
- lookback: 24 sampling steps
- lookahead: 3 sampling steps

We perform the reshaping, using the `reshape_data` function present in the `gcd_data_manipulation.py` functions collector.


```python
n_input = 24
n_out = 3
```

```python
train_x_conf_1_no_acc, train_y_conf_1_no_acc = reshape_data(
    train, n_input, n_out, input_cols_interval=(0, -1)
)
print(train_x_conf_1_no_acc.shape, train_y_conf_1_no_acc.shape)

validation_x_conf_1_no_acc, validation_y_conf_1_no_acc = reshape_data(
    validation, n_input, n_out, input_cols_interval=(0, -1)
)
print(validation_x_conf_1_no_acc.shape, validation_y_conf_1_no_acc.shape)

test_x_conf_1_no_acc, test_y_conf_1_no_acc = reshape_data(
    test, n_input, n_out, input_cols_interval=(0, -1)
)
print(test_x_conf_1_no_acc.shape, test_y_conf_1_no_acc.shape)
```

```python
type(validation_y_conf_1_no_acc)
```

## Create model

```python
def build_model(train_x, train_y, val_x, val_y):
    model = Sequential()
    model.add(LSTM(50, input_shape=(train_x.shape[1], train_x.shape[2]))) #
    model.add(Dense(train_y.shape[1]))
    model.compile(loss="mae", optimizer="adam")

    history = model.fit(
        train_x,
        train_y,
        epochs=100,
        batch_size=72,
        verbose=0,
        validation_data=(val_x, val_y),
    )
    return model, history
```

## Train model

```python
model_1, history_1 = build_model(
    train_x_conf_1_no_acc,
    train_y_conf_1_no_acc,
    validation_x_conf_1_no_acc,
    validation_y_conf_1_no_acc,
)
```

## Evaluate model

To evaluate the model, we use functions from the functions aggregator `multistepahead_model_evaluation.py`.
The main one is `walk_forward_validation`, i.e., the way we perform the inference over the test set. We look at the past 24 sampling steps, to predict the next three, using a moving window.

`walk_forward_validation` uses the function `evaluate_forecast` which computes performance metrics. The default metric is _RMSE_ but there are other ones. For every evaluation, it computes the average score, `score`, and the scores for the three predicted steps, `scores`.


### Perform test evaluation

```python
score_rmse, scores_rmse, pred_1, real_v_1 = walk_forward_validation(
    model_1, test_x_conf_1_no_acc, test_y_conf_1_no_acc, 
    test,
    scaler,
    n_out
)
(score_rmse, scores_rmse)
```
